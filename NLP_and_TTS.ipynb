```python
from gtts import gTTS
import os
from IPython.display import Audio

'''
Language settings:

English: en
Hindi: hi
French: fr
Spanish: es
German: de
Italian: it
Portuguese: pt
Japanese: ja
Chinese (Mandarin): zh-CN
'''

def text_to_speech(text, language='en', filename='output.mp3'):
    # Create a TTS object
    tts = gTTS(text=text, lang=language, slow=False)

    # Save the audio file
    tts.save(filename)

    # Return the audio file path to play it
    return filename

# Example usage
text = "Hello, this is a text-to-speech demo for the IIT Bombay humanoid project updates."
text_hindi = "नमस्ते, यह आईआईटी बॉम्बे मानवाकृति परियोजना के अपडेट के लिए एक टेक्स्ट-टू-स्पीच डेमो है।"
text_fr = "Bonjour, ceci est une démonstration de synthèse vocale pour les mises à jour du projet humanoïde de l'IIT Bombay."

audio_file = text_to_speech(text_hindi, 'hi')
Audio(audio_file)  # Play the audio file
```

```python
def read_file_and_speak(file_path, language='en'):
    with open(file_path, 'r') as file:
        text = file.read()
        audio_file = text_to_speech(text, language)
        return Audio(audio_file)

# Example usage
# Ensure to upload your text file to the Colab environment
read_file_and_speak('/content/Hello World.txt', 'hi')  # Replace with the path to your text file
```
